## Obtener HTML

```python
#Obtengo el html y lo hago un objeto beautiful soup "sopa"
r = requests.get(url_er, verify=False)
print(r.status_code)
#Si hay una respuesta mala se imprimirá esto
if not r.ok:
    print('El servidor respondió: ', r.status_code)
else:
    sopa = BeautifulSoup(r.text, 'lxml')
    #Acá además lo guardo en un archivo r.html
    with open ('r.html', 'wb') as f:
        f.write(sopa.prettify('utf-8'))
```

## Escrapear los datos que necesito

```python
#Empresa, etiqueta <th>
#Estado Financiero, etiqueta <th>
#Periodo (trimestre), etiqueta <th>
#EPS Básico == Básica por Acción Ordinaria en Operaciones Continuadas, etiqueta <td>
empresa = sopa.find('th', class_ = 'bgCeldaGuinda bordeBottom').text
estadof = sopa.find_all('th', class_ = 'bgCeldaGris txt11Normal')[0].text
periodo = sopa.find_all('th', class_ = 'bgCeldaGris txt11Normal')[2].text
epsbasico = sopa.find('td', text = 'Básica por Acción Ordinaria en Operaciones Continuadas/ ').text

print(empresa)
print(estadof)
print(periodo)
print(epsbasico)
```
